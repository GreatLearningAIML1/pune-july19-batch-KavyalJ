```python
import pandas as pd 
import numpy as np
```

```python
#mapping={'-':0}
#laliga.replace('-',0,inplace=False)
laliga=laliga.replace(
    to_replace="-",
    value=0,
    inplace=False,
)
lg=laliga
laliga.tail()
```

```python

lg['new_col'] = lg['Debut'].astype(str).str[0:4]
#laliga['new_col2']=laliga['Debut'].astype(str).str[5:]
```

```python
lg['new_col']=lg.new_col.astype(int)

debut_30_80=lg[(lg['new_col']>=1930) & (lg['new_col']<=1980)]
##laliga.drop(['new_col'],axis=1,inplace=True)
print(debut_30_80.shape)
debut_30_80[['Team','Debut']]
```

```python
#Print the list of teams which came top in terms of points
lg['new_col2']=lg.Points.astype(int)
lg.sort_values('new_col2',inplace=True,ascending=False)
lg[['Team','Points']].head(5)
```

```python
#Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same      function, find the team which has maximum and minimum goal difference. (5 points)
#Goal_diff_count = GoalsFor - GoalsAgainst
```

```python
def Goal_diff_count():
    
    lg['new_col3']=lg.GoalsFor.astype(int)
    lg['new_col4']=lg.GoalsAgainst.astype(int)
    lg['Goal_diff_count']= lg['new_col3']-lg['new_col4']
    print('\n Goal difference for all the teams:\n ')
    return lg[['Team','Goal_diff_count']]
    
    


    
    
    
```

```python
 Goal_diff_count()
```

```python
#Maximum and mininmum goal difference counts 
print(lg[['Team','Goal_diff_count']].max())
print(lg[['Team','Goal_diff_count']].min())
```

```python
#Create a new column with name “Winning Percent” and append it to the data set (5 points)
#Percentage of Winning = (GamesWon / GamesPlayed)*100
#If there are any numerical error, replace it with 0%
lg['newcol']=lg.GamesWon.astype(int)
lg['newcol1']=lg.GamesPlayed.astype(int)
lg['Winning Percent']=(lg['newcol']/lg['newcol1'])*100
lg.sort_values(by=['Winning Percent'],ascending=False)

lg['Winning Percent']
```

```python
#Winning percentage for the top 5 teams 
lg[['Team','Winning Percent']].head()
```

```python
#Group teams based on their “Best position” and print the sum of their points for all positions (5 points
#Sum of all the points for each position, so sum of points of all the teams where Best position is 1, 2 etc 
```
